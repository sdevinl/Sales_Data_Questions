```python
data = pd.read_csv('sales_data.csv')
data
```

```python
# 1)  Identify the Product line with highest total Spend 

# group by Product line, sum the spends
data_group_sum_by_product_line = data.groupby(['Product line']).sum()

# get index(product line) of max spend
highest_total_spend_product_line = data_group_sum_by_product_line['Spend'].idxmax()
highest_total_spend_product_line
```

```python
# 2) Filter the data down to the Product line identified in
# 1) and sort the data frame by ascending order of date
# and descending order of Spend. Display the first 60 rows. 

# filter data down to product line
product_line_data = data[data['Product line'] == highest_total_spend_product_line]
# sort the df by ascending order of date
product_line_data = product_line_data.sort_values(by='Date', ascending=True)
# display first 60 rows
product_line_data.head(60)
```

```python
# 3) Plot the count of orders by week for the store. 

# date timestamp libs
import time
import datetime

# plotting lib
import matplotlib.pyplot as plt


# convert 'Date' column into timestamps
data['Date_format'] = pd.to_datetime(data['Date'])

orders_by_week = data.groupby(pd.Grouper(freq='W', key='Date_format'))['Invoice ID'].count()
orders_by_week.plot()
plt.title('Number of orders per week') 
plt.ylabel('Orders')
plt.xlabel('Date')
```

```python
# 4) Compute the mean order Quantity for gender 0 and 
# for gender 1. Do you think the difference is significant and why? 

# mean quantity by gender
print(data.groupby(['Gender']).mean()['Quantity'])

# check for significance: use a T-Test
# Assuming each gender is independent and had identical variances

# Our null hypothesis: Both genders are from the same population
# Alternative hypothesis: the genders are not from the same population
from scipy.stats import ttest_ind

gender_0 = data[data['Gender'] == 0]
gender_1 = data[data['Gender'] == 1]

print('\nT-test results: ')
print(ttest_ind(gender_0['Quantity'], gender_1['Quantity']))
# Since the p-value is greater than alpha of .05, we will choose the null hypothesis.
# Gender 0 and 1 come from the same population, and there is nothing statistically significant.
```

```python
# 5) Assuming a single gender prediction was made for each customer, 
# generate a confusion matrix for predicted gender. 
# What does the confusion matrix tell you about the quality of the predictions? 

from sklearn.metrics import confusion_matrix

y_true = data['Gender']
y_pred = data['Predicted_Gender']

tn, fp, fn, tp = confusion_matrix(y_true, y_pred, labels=[0, 1]).ravel()

print('True Positive: ', tp)
print('False Positive: ', fp)
print('False Negative: ', fn)
print('True Negative: ', tn)

# 21.9% of the predictions are False Negative, a Type 2 error. 
# This is a significant amount of error in the data.
# From the confusion matrix, we can see:
# When predicting gender 1, it is incorrect 43.7% of the time.
# However, when predicting gender 0, it is correct 100% of the time. 
# The predictor is really good at prediction gender 0, and bad at predicting gender 1.
```
